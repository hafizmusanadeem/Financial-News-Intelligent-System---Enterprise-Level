# **Data has multiple tickers, so identifying the ticker with greater relevancy score is important, Moreover dropping the irrelevant ticker and getting the data from credible sources only is crucial part as well**

```python
def is_relevant_to_ticker(row, ticker, min_relevance=0.5):
    for item in row['ticker_sentiment']:
        if item['ticker'] == ticker:
            return float(item['relevance_score']) >= min_relevance
    return False

df = df[df.apply(lambda row: is_relevant_to_ticker(row, row['ticker']), axis=1)]
```
