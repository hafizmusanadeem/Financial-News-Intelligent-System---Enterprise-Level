```python
import pandas as pd
import re

# -----------------------------
# 1. Load Data
# -----------------------------

# -----------------------------
# 2. Fill Missing Values
# -----------------------------
df["headline"] = df["headline"].fillna("")
df["summary"] = df["summary"].fillna("")

# -----------------------------
# 3. Combine Text
# -----------------------------
df["text"] = (
    df["headline"].str.lower() + " " +
    df["summary"].str.lower()
)

# -----------------------------
# 4. Define Rules
# Priority matters:
# first matched rule wins
# -----------------------------
rules = {
    "earnings": [
        r"\bearnings\b",
        r"\brevenue\b",
        r"\beps\b",
        r"\bquarterly results\b",
        r"\bprofit\b",
        r"\bbeats estimates\b",
        r"\bmisses estimates\b",
        r"\bguidance\b",
    ],

    "acquisition": [
        r"\bacquire\b",
        r"\bacquires\b",
        r"\bacquired\b",
        r"\bmerger\b",
        r"\bmerge\b",
        r"\bbuyout\b",
        r"\btakeover\b",
        r"\bpurchase\b",
    ],

    "lawsuit": [
        r"\blawsuit\b",
        r"\bsued\b",
        r"\bsues\b",
        r"\blegal action\b",
        r"\bsettlement\b",
        r"\bcourt\b",
        r"\bcomplaint\b",
        r"\bfraud charges\b",
    ],

    "layoffs": [
        r"\blayoff\b",
        r"\blayoffs\b",
        r"\bjob cuts\b",
        r"\brestructure\b",
        r"\bworkforce reduction\b",
        r"\bcuts jobs\b",
        r"\bstaff reduction\b",
    ],

    "regulation": [
        r"\bsec\b",
        r"\bfda\b",
        r"\bregulator\b",
        r"\bprobe\b",
        r"\binvestigation\b",
        r"\bantitrust\b",
        r"\bapproval\b",
        r"\bban\b",
        r"\bcompliance\b",
    ],

    "product_news": [
        r"\blaunch\b",
        r"\bunveil\b",
        r"\brelease\b",
        r"\bnew product\b",
        r"\bintroduces\b",
        r"\bannounces platform\b",
        r"\brollout\b",
    ]
}

# -----------------------------
# 5. Label Function
# -----------------------------
def assign_label(text):
    for label, patterns in rules.items():
        for pattern in patterns:
            if re.search(pattern, text):
                return label
    return "other"

# -----------------------------
# 6. Apply Rules
# -----------------------------
df["event_label"] = df["text"].apply(assign_label)

```
